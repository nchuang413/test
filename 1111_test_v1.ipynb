```python
import requests

from bs4 import BeautifulSoup 

def get_list_page(url):
    """爬取1111人力銀行_軟體工程師列表頁面
    
     Parameters
    ----------
    url : str
        軟體工程師列表頁URL，
        如："https://www.1111.com.tw/job-bank/job-index.asp?d0=140202&si=1&page=1"

    Returns
    -------
    list
        頁面上所有軟體工程師列表內文URL
    """
    
    resp = requests.get(URL)
    
    soup = BeautifulSoup(resp.text, 'lxml')
    
    HOST = 'https:'
    
    for row in soup.select('h3 > a'):
        print(HOST + row['href'])
```

```python
import re
import requests

#取得列表總頁數
def get_total_page_cnt():
        
    resp = requests.get("https://www.1111.com.tw/job-bank/job-index.asp?d0=140202&si=1&page=1")
    total_page_cnt =int(re.findall('第 1 / (\d+) 頁', resp.text)[0])
    return total_page_cnt

URL = "https://www.1111.com.tw/job-bank/job-index.asp?d0=140202&si=1&col=ab&sort=desc&page={}"
LAST_PAGE = get_total_page_cnt()
PAGE = 10
DATA_DIR = '/tmp/1111bank_data'

# 爬10頁
for page_no in range(LAST_PAGE, LAST_PAGE - PAGE, -1):
    print(page_no)
    resp = requests.get(URL.format(page_no))
    
    # 把爬下來的HTML文字檔儲存到資料夾
    with open(DATA_DIR + '/index{}.html'.format(page_no), 'w') as f:
        f.write(resp.text)
```
